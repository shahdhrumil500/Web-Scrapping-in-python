```python
import requests
from bs4 import BeautifulSoup

r = requests.get("https://dir.indiamart.com/impcat/rice-mill-machinery.html")
c = r.content

soup = BeautifulSoup(c,"html.parser")
```

```python
all = soup.find_all("div",{"class":"sidsdr sid_df fww sdO"})

for item in all:
    print(item.find("div",{"class":"elps mCnmPdp"}).text)
    print(item.find("div", {"class":"sid_nm"}).text)
    print(item.find("div",{"class":"sid_prc"}).text)   
    print("\n")
```

```python
al = soup.find_all("li",{"class":"lst lst_cl mft2 img nlc"})

l=[]
for item in al:
    d={}
    s=[]
    d["Name"]= item.find("h4",{"class":"lcname"}).text
    try:
        d["Product"]= item.find("a",{"class":"fs18 ptitle"}).text
    except:
        d["Product"] = None
    d["Address"]= item.find("p",{"class":"sm clg"}).text
    d["Price"]= item.find("span",{"class":"prc cur"}).text[2:-18]
    try:
        for t in item.find("div",{"class":"desc des_p"}):
            s.append(t.text)
    except:
        d["Specification"] = None
    else:
        d["Specification"] = s
    l.append(d)

l
```

```python
len(l)
```

```python
import pandas

df = pandas.DataFrame(l)
df
```

```python
df.to_csv("Output.csv")
```
